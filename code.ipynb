```python
##LSC
import re
def lcs(X, Y, m, n): 
  
    if m == 0 or n == 0: 
           return 0; 
    elif  re.match(X[m-1],Y[n-1], re.IGNORECASE): 
        return 1 + lcs(X, Y, m-1, n-1); 
    else: 
        return max(lcs(X, Y, m, n-1), lcs(X, Y, m-1, n)); 
```

```python
import codecs
def build_word_vector_matrix(vector_file):
    embedding_index = {}
    with codecs.open(vector_file, 'r', 'utf-8') as f:
        for i, line in enumerate(f):
            sr = line.split()
            if(len(sr)<26):
                continue
            word = sr[0]
            embedding = np.asarray(sr[1:], dtype='float32')
            embedding_index[word] = embedding
    return embedding_index
# Replace the path here to point to the glove.6B.50d.txt vectors file on your system
words = build_word_vector_matrix('/home/pranab/Documents/machine_larning/glove.6B.50d.txt')

word_to_vec_map= build_word_vector_matrix('/home/pranab/Documents/machine_larning/glove.6B.50d.txt')
#print(word_to_vec_map)
```

```python
def cosine_similarity(u, v):
    """
    Cosine similarity reflects the degree of similariy between u and v
        
    Arguments:
        u -- a word vector of shape (n,)          
        v -- a word vector of shape (n,)

    Returns:
        cosine_similarity -- the cosine similarity between u and v defined by the formula above.
    """
    
    distance = 0.0
    
    ### START CODE HERE ###
    # Compute the dot product between u and v (1 line)
    dot = np.dot(u, v)
    # Compute the L2 norm of u (1 line)
    norm_u = np.sqrt(np.sum(u * u))
    
    # Compute the L2 norm of v (1 line)
    norm_v = np.sqrt(np.sum(v * v))
    # Compute the cosine similarity defined by formula (1) (1 line)
    cosine_similarity = dot / (norm_u * norm_v)
    ### END CODE HERE ###
    
    return cosine_similarity
```
